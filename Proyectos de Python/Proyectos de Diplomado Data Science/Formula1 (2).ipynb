```python
#importamos las librerias que se utilizarán
import pandas as pd
import requests
```

```python
#creamos la ruta del archivo en una variable para llmarla mas tarde, esto (personalmente) me hace mas fácil la lectura
ruta_archivo = 'https://github.com/JFRo57/Movies-Dataset/raw/main/racesOpen.csv'
```

```python
#importamos y creamos el dataframe 
df= pd.read_csv(ruta_archivo, encoding="ISO-8859-1", index_col=0)
```

```python
#Haremos una vista rápida para observar los datos que contiene
df.head(5)
```

```python
#Haremos una copia del Dataframe para poder volver por si ocurre un error (según lo recomendado en el curso)
df1=df.copy()
#También aprovecharemos para ver cuantos archivos existen en el df
df1.tail(3)
```

```python
#Antes de empezar a procesar datos, necesitamos saber que cual es el tamño de la información que veremos a continuación
df1.shape  #Esto nos dice cuantas columnas y filas existen 
```

```python
#Ahora veremos los tipos de datos que existen en la tabla
df1.dtypes
```

```python
#Veamos a que se refiere cada columna:
#raceId: Identificación única de la carrera.
#year: El año en el que tuvo lugar la carrera.
#round: En que orden se llevó a cabo la carrera, si es 1 es la primera de la temporada, así sucesivamente.
#circuitId: Identificación única del circuito en el que se llevó a cabo la carrera.
#name: El nombre del Gran Premio o carrera (por ejemplo, "Australian Grand Prix" o "Malaysian Grand Prix").
#date: La fecha en la que se celebró la carrera (en formato año-mes-día).
#time: La hora de inicio de la carrera (en formato hora:minuto:segundo).
#url: Un enlace URL del que se obtuvo la información
```

```python
df1['year'] = df1['year'].astype('int64')
df1['round'] = df1['round'].astype('int64')
df1['circuitId'] = df1['circuitId'].astype('int64')
df1['date'] = pd.to_datetime(df['date'])
df1['url'] = df1['url'].astype('str')
df1['time'] = pd.to_datetime(df1['time'], format='%H:%M:%S').dt.time
```

```python
df1.dtypes
```

```python
df1.head(5)
```

```python
#Eliminaremos las columnas que no son necesarias el presente proyecto 
df1=df1.drop(columns=['time','url']) 
```

```python
#La vista del archivo se vería de la siguiente forma
df1.head(5)
```

```python
#En este punto no se notó mas información que se necesite o que deba ser omitida
#cambiaremos los nombre únicamente con fines didácticos y para usar un diccionario el cual, como su nombre lo indica...
#contiene una variable (key) y un valor; esta estructura puede redimensionarse. En este caso las key son los nombres..
#de las columnas y les asignaremos un valor o descripción con nombres comprensibles
df1.rename(columns={
'year':'Año',
'name':'Gran Premio',
'date':'Fecha'},inplace = True)
```

```python
#Como no usaremos ni la posición en latitud y longitud, y tampoco la url, las eliminaremos del df, 
#al parecer circuitRef es un ID sin embargo como no está relacionada con el df inicial la omitiremos
df3_dropped=df3.drop(columns=['url','lat','lng','circuitRef']) 
```

```python
#Como vemos tenemos datos NaN que no sirven para el análisis, así que procederemos a elimianrlos 
df4=df3_dropped.dropna(axis=1, how='any')
```

```python
#Cambiemos entonces los nombres de las columnas a unos mas comprensibles
df4.rename(columns={
'name':'Pista',
'location':'Ciudad',
'country':'País'},inplace = True)
```

```python
#Veamos ahora el nuevo dataframe
df4.head(5)
```

```python
result = df1.merge(df4, left_on='circuitId', right_on='circuitId',suffixes=('_left', '_right'))
result.tail(7)
```
